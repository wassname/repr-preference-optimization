```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from reprpo.data.datamodule import PrefDataModule
from transformers import AutoTokenizer
from argparse import ArgumentParser, Namespace

model_name = "wassname/llama-3-2-1b-sft"

tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
```

```python
datasets = [
    # set 1
    # FIXME, check truncation on these, should ideally be none
    "code",
    "math",
    "alpaca_mmlu",
    # "code_easy",
    "cooking",

    "alpaca_low_quality",
    "math_easy",

    # set 2
    "code_easy",
    "us_history",
    "change_my_view",
    "raven_matrices",

    # set 3
    "ranking_logic_easy",
    "shp_low_quality",
    "pursue_goals",
    "creative_writing",
    "alpaca_easy",
    "arc_easy",
    "us_history_textbook",
    "alpaca_chat",
    "raven_easy",
    "code_low_quality",
    "alpaca_short",
]
```

```python
for ds_name in datasets:
    args = Namespace(
        dataset=ds_name,
        verbose=0,
        max_prompt_length=1024,
        max_length=2048,
        batch_size=1)
    dm = PrefDataModule(args, tokenizer=tokenizer)
    dm.setup()

    max_sizes = np.array(
        [sum(x) for x in dm.train_ds['rejected_mask']] + 
        [sum(x) for x in dm.train_ds['chosen_mask']]
    )
    prompt_sizes = np.array([sum(x) for x in dm.train_ds['prompt_mask']])
    logger.info(f"""{ds_name} max_sizes
- Max : {max_sizes.max()}, 
- 99th: {np.percentile(max_sizes, 99)},
- 90th: {np.percentile(max_sizes, 90)}, 
- 75th: {np.percentile(max_sizes, 75)}

prompt_sizes
- Max : {prompt_sizes.max()},
- 99th: {np.percentile(prompt_sizes, 99)},
- 90th: {np.percentile(prompt_sizes, 90)},
- 75th: {np.percentile(prompt_sizes, 75)}""")

    plt.figure(figsize=(3, 2.5))
    plt.hist(max_sizes, bins=100, alpha=0.5, label="Max Sizes")
    plt.title(f"Dataset: {ds_name}")
    plt.xlabel("Size of completions")
    plt.hist(prompt_sizes, bins=100, alpha=0.5, label="Prompt Sizes")
    plt.ylabel("Count")
    plt.legend()
    plt.show()
```
